# Introducción a Matplotlib gráfico de líneas y dispersión

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

month = np.array(['E', 'F', 'M', 'A', 'Ma'])
sales = np.array([20, 25, 30, 28, 35])

#Configurar el tamaño del grafico
plt.figure(figsize=(8,6))

#Crear garfico
plt.plot(month, sales, marker='o', color='b')

plt.title('Ventas mensuales de un producto')
plt.xlabel('Meses')
plt.ylabel('Ventas')

plt.show()
```

## Grafico de dispersión

```python
import matplotlib.pyplot as plt

#Configurar el tamaño del grafico
plt.figure(figsize=(8,6)) # 8 de ancho y 6 de alto

hours = [2,3,4,5,6,7,8]
exam = [55, 60, 65, 70, 75, 80, 85]

# Crear grafico de dispersion
plt.scatter(hours, exam, color='green')

plt.title('Relación entre horas estudiadas y el puntaje') # Titulo
plt.xlabel('Horas')# Etiqueta del eje x
plt.ylabel('Puntajes')# Etiqueta del eje y

plt.show()
```

# Personalización

```python
#Configurar el tamaño del grafico
plt.figure(figsize=(8,6)) # 8 de ancho y 6 de alto

hours = [2,3,4,5,6,7,8]
exam_scores_student_1 = [55, 60, 65, 70, 75, 80, 85]
exam_scores_student_2 = [50, 58, 63, 69, 74, 78, 83]

# Crear grafico de dispersion de dos estudiantes
#Estudiante 1
plt.scatter(hours, exam_scores_student_1, marker='o', color='green', label='Estudiante 1')

#Estudiante 2
plt.scatter(hours, exam_scores_student_2, marker='s', color='red', label='Estudiante 2')

plt.title('Relación entre horas estudiadas y el puntaje de dos estudiantes') # Titulo
plt.xlabel('Horas')# Etiqueta del eje x
plt.ylabel('Puntajes')# Etiqueta del eje y

plt.show()
```

## Con grafico de lineas

```python
#Configurar el tamaño del grafico
plt.figure(figsize=(8,6)) # 8 de ancho y 6 de alto

hours = [2,3,4,5,6,7,8]
exam_scores_student_1 = [55, 60, 65, 70, 75, 80, 85]
exam_scores_student_2 = [50, 58, 63, 69, 74, 78, 83]

# Crear grafico de dispersion de dos estudiantes
#Estudiante 1
plt.plot(hours, exam_scores_student_1, marker='o', color='green', linestyle='-', linewidth = 2, label='Estudiante 1')

#Estudiante 2
plt.plot(hours, exam_scores_student_2, marker='s', color='red', linestyle='--',linewidth = 2, label='Estudiante 2')

plt.title('Relación entre horas estudiadas y el puntaje de dos estudiantes') # Titulo
plt.xlabel('Horas')# Etiqueta del eje x
plt.ylabel('Puntajes')# Etiqueta del eje y

plt.show()
```

## Ejemplo

```python
notas = [4.0, 3.6, 7.0, 8.0, 9.3, 4.5, 6.0]
students = ['Juan', 'Andres', 'Pedro', 'Maria', 'Ana', 'Jose', 'Luis']

plt.figure(figsize=(8,6))
plt.bar(students, notas, color='blue')
plt.title('Notas de los alumnos')
plt.xlabel('Alumnos')
plt.ylabel('Notas')
plt.show()
```

## Barras y Pastel

```python
plt.figure(figsize=(8,6))

categories = ['Producto A', 'Producto B', 'Producto C']
sales = [120,150,90]

#Creacion del grafico de barras en verticales
plt.bar(categories,sales, color='skyblue', label='Ventas mensuales')

#Anotación con flecah para destacar un punto específico
plt.annotate('maximo de ventas', xy=('Producto B', 150),xytext=('Producto C', 160), arrowprops=dict(facecolor='black', shrink=0.05))


plt.title('Ventas de productos en un mes')
plt.xlabel('Productos')
plt.ylabel('Ventas')

#plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))

categories = ['Producto A', 'Producto B', 'Producto C']
sales = [120,150,90]

#Creacion del grafico de barras en verticales
plt.barh(categories,sales, color='skyblue', label='Ventas mensuales')

plt.title('Ventas de productos en un mes')
plt.ylabel('Productos')
plt.xlabel('Ventas')

#plt.legend()
plt.show()
```

## Pastel

```python
plt.figure(figsize=(8,6))

products = ['Producto A', 'Producto B', 'Producto C']
markert_share = [50,35,15]

#Creacion del diagrama de pastel
plt.pie(markert_share, labels=products, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'orange'])

#Titulo
plt.title('Ventas de productos en un mes')

# Asegurar que el grafico sea un circulo
plt.axis('equal')

plt.show()
```
