# Hitogramas y boxplot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = np.random.normal(170, 10, 200) # Generate random data
```

```python
plt.hist(data, color = 'salmon', bins=20, edgecolor = 'black', alpha=0.6) # Create histogram
plt.title('Distribución de altura')
plt.xlabel('Altura (cm)') # X-axis label
plt.ylabel('Densidad')
plt.show()
```

```python
np.random.seed(0) # Set seed for reproducibility
ages = [np.random.normal(30, 5, 100),
        np.random.normal(40, 5, 100),
        np.random.normal(35, 5, 100)]
```

```python
plt.boxplot(ages, patch_artist=True, notch=True, vert=True, label=('Grupo 1', 'Grupo 2', 'Grupo 3')) # Create boxplot
plt.title('Distribución de edades')
plt.xlabel('Grupos')
plt.ylabel('Edad (años)') # Y-axis label
plt.show()
```
